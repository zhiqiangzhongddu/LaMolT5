```python
import matplotlib.pyplot as plt
```

```python
# Molecule Generation
data = {
    'LaMolT5-Small': {'x': 7.7e6, 'y': 0.594, 'color': 'red'},
    'LaMolT5-Base': {'x': 7.5e8, 'y': 0.596, 'color': 'green'},
    'LaMolT5-Large': {'x': 8e8, 'y': 0.593, 'color': 'yellow'},
    # 
    'MolT5-Small': {'x': 7.7e6, 'y': 0.482, 'color': 'brown'},
    'MolT5-Base': {'x': 7.5e8, 'y': 0.496, 'color': 'pink'},
    'MolT5-Large': {'x': 8e8, 'y': 0.554, 'color': 'grey'},
    # 
    'MolReGPT': {'x': 2e10, 'y': 0.593, 'color': 'grey'},
    'MolT5-Large-HV': {'x': 7.7e8, 'y': 0.59, 'color': 'blue'},
    'TGM-DLM w/o corr': {'x': 1.8e8, 'y': 0.589, 'color': 'purple'},   
}

plt.figure(figsize=(8, 6))
for key, item in data.items():
    plt.scatter(item['x'], item['y'], color=item['color'], label=key)
    if key == 'LaMolT5-Large':
        plt.annotate(
            key,
            (item['x'], item['y']),
            textcoords="offset points",
            xytext=(7, -3),
            ha='left'
        )
    elif key == 'MolT5-Large-HV':
        plt.annotate(
            key,
            (item['x'], item['y']),
            textcoords="offset points",
            xytext=(7, -7),
            ha='left'
        )
    elif key == 'LaMolT5-Small' or key == 'MolT5-Small':
        plt.annotate(
            key,
            (item['x'], item['y']),
            textcoords="offset points",
            xytext=(7, -3),
            ha='left'
        )
    elif key == 'MolReGPT':
        plt.annotate(
            key,
            (item['x'], item['y']),
            textcoords="offset points",
            xytext=(-5, 0),
            ha='right'
        )
    else:
        plt.annotate(
            key,
            (item['x'], item['y']),
            textcoords="offset points",
            xytext=(5, 5),
            ha='center'
        )

plt.xscale('log')
plt.xlabel('Number of Parameters')
plt.ylabel('Text2Mol')

# save image
plt.savefig('../../output/figures/molecule_generation.pdf', bbox_inches="tight")

# Show plot
plt.show()
```
