```python
import pandas as pd
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")

data = pd.DataFrame(
    {
        "caption": ["Original", "Original", "Original", "Original", "Original", "Original", "Original", "Original", "Original",
                    "ChatGPT + Gemini", "ChatGPT + Gemini", "ChatGPT + Gemini", "ChatGPT + Gemini", "ChatGPT + Gemini", "ChatGPT + Gemini", "ChatGPT + Gemini", "ChatGPT + Gemini", "ChatGPT + Gemini", 
                    "Llama 2-70b + Llama 3-70b", "Llama 2-70b + Llama 3-70b", "Llama 2-70b + Llama 3-70b", "Llama 2-70b + Llama 3-70b", "Llama 2-70b + Llama 3-70b", "Llama 2-70b + Llama 3-70b", "Llama 2-70b + Llama 3-70b", "Llama 2-70b + Llama 3-70b", "Llama 2-70b + Llama 3-70b"],
        "metrix": ["BLEU", "Exact", "Lev.", "MAC.", "RDK.", "Mor.", "FCD", "Text2Mol", "Val.",
                   "BLEU", "Exact", "Lev.", "MAC.", "RDK.", "Mor.", "FCD", "Text2Mol", "Val.",
                   "BLEU", "Exact", "Lev.", "MAC.", "RDK.", "Mor.", "FCD", "Text2Mol", "Val.",],
        "performance": [0.755, 0.079, 25.988, 0.703, 0.568, 0.517, 1.35, 0.482, 0.721,
                        0.852, 0.287, 16.009, 0.891, 0.805, 0.741, 0.384, 0.594, 0.950,
                        0.852, 0.287, 16.009, 0.891, 0.805, 0.741, 0.384, 0.594, 0.950,],
    }
)

standard = 0.755, 0.079, 25.988, 0.703, 0.568, 0.517, 1.35, 0.482, 0.721
for idx, value in enumerate(data.performance):
    data.loc[idx, "performance"] = data.performance[idx] / standard[idx % 9]

# Draw a nested barplot by species and sex
g = sns.catplot(
    data=data, kind="bar",
    x="metrix", y="performance", hue="caption",
    errorbar="sd", palette="dark", alpha=.6, height=6
)

g.despine(left=True)
g.set_axis_labels("", "Relative Score")
g.legend.set_bbox_to_anchor((.6, .5))
g.legend.set_title("")

# Show plot
g.fig.tight_layout()
g.fig.subplots_adjust(right=0.8)
g.savefig('../../output/figures/different_captions.pdf', bbox_inches="tight")
```
